```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import glob
```

```python
results = [pd.read_csv(path) for path in glob.glob("../logs/*csv")]
results = pd.concat(results)
# Clean dataframe
results = results.apply(lambda x: x.apply(lambda y: y.split(' ')[1][:-1]))
results
```

```python
ids = results['experiment_id'].to_numpy()
ids = np.array([id[1:-1] for id in ids]).astype(int)
```

```python
ids.sort()
```

```python
ids_complete = np.arange(1,11401)
ids_complete
```

```python
err_ids = [id for id in ids_complete if id not in ids]
```

```python
len(err_ids)
```

```python
err_ids
```

```python
unfinished_file = open("unfinished.txt", "a")
# f.write("Now the file has more content!")

with open('../code/nohup.out') as f:
    lines = f.readlines()
    for line in lines:
        if 'GPU(' in line:
            exp_id = int(line.split(' /')[0].split('(')[1])
            command = line.split(': ')[1].rstrip()
            if exp_id in err_ids:
                unfinished_file.write(command + '\n')
unfinished_file.close()
```
